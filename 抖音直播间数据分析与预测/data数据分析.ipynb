```python
new_col = ['del','自然流量详情','付费流量详情','UV价值','新增评论数','时间','新增粉丝数','离开人数','新加团人数','成交金额','在线人数','进入人数','成交订单数','成交人数','企业唯一标识']
data.columns = new_col
data.drop('del',axis=1,inplace=True)
data.head()
```

```python
# 将一列切分出多个独立的列
def splitNewCols(data,colname,dic_,n):
    for value_str in data[colname]:
        lis_ = value_str.split(',')
        a = 0
        for element in lis_:
            name = element.split(':')[0][2:-1]
            if a==n:
                value = int(element.split(':')[1][:-1])
            else:
                value = int(element.split(':')[1])
            dic_[name].append(value)
            a += 1
    return dic_
```

```python
dic1 = {'短视频引流':[],'抖音商城推荐':[],'直播推荐':[],'搜索':[],'个人主页&店铺&橱窗':[],'其他':[],'头条西瓜':[],'活动页':[],'全部自然流量':[],'关注':[]}
dic2 = {'千川PC版': [], '其他广告': [], '千川品牌广告': [], '全部付费流量': [], '品牌广告': [], '小店随心推': []}
data_dic1 = splitNewCols(data,'自然流量详情',dic1,9)
data_dic2 = splitNewCols(data,'付费流量详情',dic2,5)
data_dic1 = pd.DataFrame(data_dic1)
data_dic2 = pd.DataFrame(data_dic2)
print(data.shape)
new_data = pd.concat([data,data_dic1,data_dic2],axis=1)
new_data.drop(['自然流量详情','付费流量详情'],axis=1,inplace=True)
print(new_data.shape)
```

```python
new_data['时间'] = pd.to_datetime(new_data['时间'])
new_data['月'] = new_data['时间'].dt.month
new_data['天'] = new_data['时间'].dt.day
new_data['小时'] = new_data['时间'].dt.hour
```
